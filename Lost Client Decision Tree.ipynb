```python
import pandas as pd
import numpy as np
```

```python
lc = pd.read_csv(r'C:\Users\Chr041921\Downloads\CQ_LOSTCLIENTTRACKER1.csv')
```

```python
lc.head()
```

```python
lc.dtypes
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(lc.drop('LOSTCLIENT',1), lc['LOSTCLIENT'], test_size=0.2, random_state=4811)
```

```python
X_train.info()
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(random_state=4811)
tree.fit(X_train, y_train)

train_data_score=tree.score(X_train, y_train)
test_data_score=tree.score(X_test, y_test)

print("the score of the decision tree in the training data is " + str(train_data_score))
print("the score of the decision tree in the testing data is " + str(test_data_score))
```

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=5, random_state=4811)
forest.fit(X_train,y_train)

train_data_score_rf=forest.score(X_train, y_train)
test_data_score_rf=forest.score(X_test, y_test)

print("the score of the random forest tree in the training data is " + str(train_data_score_rf))
print("the score of the random forest tree in the testing data is " + str(test_data_score_rf))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbtree=GradientBoostingClassifier(learning_rate=0.2, max_depth=2, random_state=4811)
gbtree.fit(X_train,y_train)

train_data_score_gb=gbtree.score(X_train, y_train)
test_data_score_gb=gbtree.score(X_test, y_test)

print("the score of the gradient boosting tree in the training data is " + str(train_data_score_gb))
print("the score of the gradient boosting tree in the testing data is " + str(test_data_score_gb))
```

```python
raw_data=lc.sample(frac=0.25, random_state=4811)
raw_data.head()
raw_data_feat=raw_data.drop(columns=['LOSTCLIENT'])
raw_data_feat.head()
decision_tree_pred=tree.predict(raw_data_feat)
random_forest_pred=forest.predict(raw_data_feat)
gradient_boosting_pred=gbtree.predict(raw_data_feat)
raw_data['decision_tree_prediction']=decision_tree_pred
raw_data['random_forest_prediction']=random_forest_pred
raw_data['gradient_boosting_pred']=gradient_boosting_pred

raw_data['avg']=(raw_data['decision_tree_prediction']+raw_data['random_forest_prediction']+raw_data['gradient_boosting_pred'])/3
raw_data['majority_pred']=np.where(raw_data['avg']>=2/3,1,0)
raw_data.head()
```

```python
raw_data.to_excel("Lost_Client_AI_Predictions.xlsx")
```
